# GLUE Dataset Explained

## What is GLUE?

**GLUE (General Language Understanding Evaluation)** is a benchmark collection of datasets used to evaluate how well NLP models understand human language.

It is **not a single dataset**. Instead, it is a collection of multiple datasets, where each dataset represents a different Natural Language Processing (NLP) task.

---

## Why Was GLUE Created?

Suppose someone says:

> "My NLP model is very good."

A natural question would be:

- Good at sentiment analysis?
- Good at text similarity?
- Good at inference?
- Good at question matching?

To provide a standardized way to evaluate language understanding, researchers created **GLUE**.

Models such as BERT are often evaluated using GLUE tasks.

---

## GLUE Tasks

### 1. SST-2 (Sentiment Analysis)

Determine whether a sentence expresses a positive or negative sentiment.

**Example:**

Input:

```text
This movie is amazing.
```

Output:

```text
Positive (1)
```

---

### 2. MRPC (Microsoft Research Paraphrase Corpus)

Determine whether two sentences have the same meaning.

**Example:**

Sentence 1:

```text
The company released a new phone.
```

Sentence 2:

```text
A new smartphone was launched by the company.
```

Output:

```text
Yes (1)
```

---

### 3. QQP (Quora Question Pairs)

Determine whether two questions are duplicates.

**Example:**

Question 1:

```text
How can I learn Python?
```

Question 2:

```text
What is the best way to learn Python?
```

Output:

```text
Duplicate (1)
```

---

### 4. MNLI (Multi-Genre Natural Language Inference)

Predict the relationship between a sentence and a hypothesis.

**Example:**

Premise:

```text
A man is playing cricket.
```

Hypothesis:

```text
Someone is playing a sport.
```

Output:

```text
Entailment
```

Possible labels:

- Entailment
- Contradiction
- Neutral

---

### 5. CoLA (Corpus of Linguistic Acceptability)

Determine whether a sentence is grammatically correct.

**Example:**

Input:

```text
The cat sits on the mat.
```

Output:

```text
Acceptable
```

---

### 6. RTE (Recognizing Textual Entailment)

Determine whether one sentence logically follows from another.

**Example:**

Text:

```text
The president signed the bill.
```

Hypothesis:

```text
The bill became law.
```

Output:

```text
Entailment / Not Entailment
```

---

### 7. STS-B (Semantic Textual Similarity Benchmark)

Measure how similar two sentences are.

**Example:**

Sentence 1:

```text
A man is riding a bicycle.
```

Sentence 2:

```text
A person is cycling.
```

Output:

```text
Similarity Score: 4.8 / 5
```

---

### 8. WNLI (Winograd Natural Language Inference)

Tests a model's ability to understand pronouns and sentence context.

**Example:**

```text
The trophy doesn't fit in the suitcase because it is too large.
```

Question:

```text
What does "it" refer to?
```

Answer:

```text
The trophy
```

## GLUE vs WMT14

| Feature | GLUE | WMT14 |
|----------|----------|----------|
| Type | Benchmark Collection | Translation Dataset |
| Purpose | Language Understanding | Machine Translation |
| Tasks | Multiple NLP Tasks | Hindi ↔ English Translation |
| Output | Labels/Scores | Translated Text |
| Example Model | BERT | T5, MarianMT, mBART |

---

### Key Takeaway

- **GLUE** is a benchmark containing multiple datasets for evaluating language understanding.
- **WMT14** is a dataset specifically designed for machine translation.
- GLUE tasks typically output **labels or scores**, while WMT14 outputs **translated sentences**.
