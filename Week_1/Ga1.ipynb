## Que 3

`ds.cache_files`: represents the size of the downloaded dataset stored locally in the cache.

````python
from datasets import load_dataset
import os

ds = load_dataset("ai4bharat/naamapadam", "ta")

total_size = 0

for split in ds.cache_files:
    for file_info in ds.cache_files[split]:
        size_mb = os.path.getsize(file_info["filename"]) / (1024 * 1024)
        total_size += size_mb

print(f"Total dataset size: {total_size:.2f} MB")
````

## Quick Summary

| Attribute | What it Represents | When Measured |
|------------|-------------------|----------------|
| `builder.info.download_size` | Size of the files downloaded from the dataset source (compressed/raw files) | Before processing |
| `builder.info.dataset_size` | Size of the processed dataset stored in Apache Arrow format | After processing |
| `os.path.getsize()` on `ds.cache_files` | Actual size of the cached dataset files stored on your machine | After download and processing |

### In Simple Terms

```text
Source Files (download_size)
            ↓
      Processing
            ↓
Arrow Dataset Files (dataset_size)
            ↓
Stored in Cache Directory
            ↓
Actual File Size on Disk (os.path.getsize)

## Que 8


### Why Dataset Modifications Increase Disk Usage

Hugging Face datasets are **immutable**. Operations such as:

```python
ds = ds.map(...)
ds = ds.filter(...)
ds = ds.remove_columns(...)
ds = ds.rename_column(...)
```

do **not** modify the existing Apache Arrow file in place.

Instead, the library:

1. Reads the existing cached dataset.
2. Creates a **new Arrow cache file** containing the modified dataset.
3. Keeps the old cache file for reproducibility and caching purposes.

As a result, the cache directory may contain multiple versions of the dataset, causing disk usage to increase over time.

#### Example

Initial dataset:

```text
cache/
├── original.arrow
```

After:

```python
ds = ds.map(add_token_col)
```

Cache directory:

```text
cache/
├── original.arrow
├── modified.arrow
```

After another transformation:

```python
ds = ds.filter(...)
```

Cache directory:

```text
cache/
├── original.arrow
├── modified.arrow
├── filtered.arrow
```

#### Conclusion

Each dataset transformation typically creates a new cache file rather than modifying the existing one. Therefore, repeated operations such as `map()`, `filter()`, and `remove_columns()` can gradually increase the disk space consumed by the Hugging Face cache.
