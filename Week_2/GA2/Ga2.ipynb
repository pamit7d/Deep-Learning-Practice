````python
from tokenizers import Tokenizer
from tokenizers.normalizers import Lowercase
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.models import BPE
````

````python
model=BPE(unk_token='UKN')

tokenizer = Tokenizer(model)
````

````python
tokenizer.normalizer = Lowercase()
tokenizer.pre_tokenizer = Whitespace()
````

````python
from tokenizers.trainers import BpeTrainer

trainer = BpeTrainer(vocab_size=5000, special_tokens= ['GO','UNK','PAD','EOS'], continuing_subword_prefix='##')
````

````python
def get_sentence(batch_size = 10000):
    for idx in range(0, len(ds), batch_size):
        yield ds['text'][idx : (idx + batch_size)]
        
````

````python
tokenizer.train_from_iterator(get_sentence(batch_size=10000), trainer=trainer, length=len(ds))
````

````python
import os

if not os.path.exists('ga_model'):
    os.mkdir('ga_model')
    
tokenizer.model.save('ga_model', prefix='GA')
````

````python
sample = 'SEBI study finds 93% of individual F&O traders made losses between FY22 and FY24.'
````

````python
encoding = tokenizer.encode(sample)

print(encoding)
````

````python
from datasets import load_dataset

imdb_dataset = load_dataset("stanfordnlp/imdb")
print(imdb_dataset)
````

- Droping the unsupervised feature

````python
from datasets import concatenate_datasets

all_ds = concatenate_datasets([
    imdb_dataset["train"],
    imdb_dataset["test"]
])

print(len(all_ds))

all_text = all_ds["text"]
````

````python
def count_tokens(tok, texts):
    total = 0
    for text in texts:
        enc = tok.encode(text)
        total += len(enc.ids) if hasattr(enc, 'ids') else len(enc)
    return total
````

### Line-by-Line Explanation

```python
total = 0
```

- Counter to accumulate the token count across all texts.

---

```python
for text in texts:
```

- Loop over each of the 50,000 IMDB reviews one by one.

---

```python
enc = tok.encode(text)
```

- Tokenize one review.
- Different libraries return different types:

  - **`tokenizers` library (custom/Hugging Face Tokenizers)** → returns an **`Encoding` object** → access token IDs using `enc.ids`.
  - 
  - **`transformers` library (BERT, GPT-2, etc.)** → returns a **plain list of token IDs** directly.

---

```python
hasattr(enc, "ids")
```

- Checks whether the returned object has an `.ids` attribute.

  - **If `True`** → the tokenizer is from the **`tokenizers`** library → use:

    ```python
    len(enc.ids)
    ```

  - **If `False`** → the tokenizer is from the **`transformers`** library → use:

    ```python
    len(enc)
    ```

---

```python
total += ...
```

- Adds the token count of the current review to the running total.

---

```python
return total
```

- Returns the total number of tokens across all 50,000 IMDB reviews.

---

### Summary

The `hasattr(enc, "ids")` check simply handles the difference between the **`tokenizers`** and **`transformers`** libraries, allowing the same function to work with both types of tokenizers.

````python
# Count tokens for all 4 tokenizers
tokens_hopper = count_tokens(trained_tokenizer, all_text)
tokens_custom = count_tokens(tokenizer, all_text)
tokens_bert   = count_tokens(bert_tokenizer, all_text)
tokens_gpt2   = count_tokens(gpt2_tokenizer, all_text)

print("Hopper :", tokens_hopper)
print("Custom :", tokens_custom)
print("BERT   :", tokens_bert)
print("GPT2   :", tokens_gpt2)

# Sort in ascending order
results = {
    "1. Custom 32K": tokens_custom,
    "2. BERT"      : tokens_bert,
    "3. GPT2"      : tokens_gpt2,
    "4. Hopper"    : tokens_hopper,
}

sorted_results = sorted(results.items(), key=lambda x: x[1])
print("\nAscending order:")
for name, count in sorted_results:
    print(f"  {name}: {count}")
````
